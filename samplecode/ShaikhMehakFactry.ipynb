```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\maintenance_data\maintenance_data.csv")
```

```python
df.shape
```

```python
df.columns
```

```python
df.head().T
```

```python
a=['broken','team','provider']
b=['lifetime','pressureInd','moistureInd','temperatureInd']
```

```python
for i in a:
    sns.countplot(x=i,data=df)
    plt.show()
    print(df[i].describe())
```

```python
for j in b:
    sns.distplot(df[j],kde=False)
    plt.show()
    print(df[j].describe())
    print(df[j].skew())
    
```

```python
sns.countplot(x="provider",hue="broken",data=df)
plt.show()
```

```python
sns.countplot(x="team",hue="provider",data=df)
plt.show()
```
